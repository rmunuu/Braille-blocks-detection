```python
import torch
import cv2
from matplotlib import pyplot as plt
from pathlib import Path
import pathlib
temp = pathlib.PosixPath
pathlib.PosixPath = pathlib.WindowsPath

# Load the YOLOv5 model
model = torch.hub.load("ultralytics/yolov5", "custom", path="best.pt", force_reload=True)

# Define a function to perform detection
def detect(image_path):
    # Load the image
    img = cv2.imread(image_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Perform inference
    results = model(img)

    # Print results
    results.print()  # Print results to the console
    results.show()   # Display results

    # Save results
    # results.save(save_dir=Path(image_path).parent / 'results')

    # Display the image with detections
    plt.imshow(img)
    plt.show()

# Example usage
detect("C:/Users/choid/Downloads/인도보행 영상/바운딩박스/Bbox_27_new/Bbox_2205/ZED3_KSC_080853_L_P024541.png")
```
